```python
vectorizer = CountVectorizer(stop_words='english')
doc_word = vectorizer.fit_transform(tweets)
doc_word.shape
```

```python
vectorizer.get_feature_names()[-2000]
```

```python
doc_term=pd.DataFrame(doc_word.toarray(), index=tweets, columns=vectorizer.get_feature_names())
doc_term.head(10)
```

```python
lsa = TruncatedSVD(5)
doc_topic = lsa.fit_transform(doc_word)
lsa.explained_variance_ratio_
```

```python
topic_word = pd.DataFrame(lsa.components_.round(3),
             index = ["component_1","component_2", "component_3", "component_4",  "component_5"],
             columns = vectorizer.get_feature_names())
topic_word
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(lsa, vectorizer.get_feature_names(), 25)
```
