```python
with open('twitter_usernames.txt', 'r') as filehandle:
    twitter_usernames=json.load(filehandle)
```

```python
twitter_usernames
```

##### USE USERNAMES TO GRAB TWEETS

```python
# Configure

for user in twitter_usernames: 
    try: 
        c = twint.Config()
        c.Username = user
        c.Since = "2020-12-01"
        c.Until = "2021-02-17"
        c.Store_csv = True
        c.Output = "econ_tweets"
        twint.run.Search(c)
    except: 
        pass
```
