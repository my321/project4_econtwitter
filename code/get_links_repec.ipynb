```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
url = 'https://ideas.repec.org/i/etwitter.html' 

response = requests.get(url)
```

```python
response.status_code
```

```python

response.text[:1000]
```

```python
page=response.text
```

```python
soup=BeautifulSoup(page, 'lxml')
```

```python
print(soup.prettify())
```

##### GET LINKS FROM TABLE ON THE REPEC WEBSITE 

```python
repec_links=[]
for i in range(1,27):
    for item in soup.find_all('table')[i].find_all('a'): 
        repec_links.append(item.get('href'))
```

```python
len(repec_links)
```

```python
with open('repec_links.txt', 'w') as filehandle:
    json.dump(repec_links, filehandle)
```
