```python
import pandas as pd 
import re
from langdetect import detect
import json 
```

```python
df=pd.read_csv('econ_tweets/tweets.csv') 
 
```

```python
df['tweet']
```

```python
tweets=df['tweet'].tolist()
```

## CLEAN TWEETS OF PUNCTUATION
<p> Remove @, emoticons, emojis, exclamation, periods, question </p>

##### Remove @

```python
tweets_remove_at=[tweet.replace('@','') for tweet in tweets]
```

```python
tweets_remove_at
```

##### Remove http links

```python
tweets_no_links=[re.sub(r'http\S+', '', tweet) for tweet in tweets_remove_at]
```

```python
tweets_no_links
```

##### Remove punctuation

```python
tweets_no_punct=[re.sub(r'[^\w\s]', '', tweet) for tweet in tweets_no_links]
```

```python
tweets_no_punct
```

##### Remove non-ASCII words from tweets 

```python
tweets_no_ascii=[re.sub(r'[^\x00-\x7F]','', tweet) for tweet in tweets_no_punct]
```

```python
tweets_no_ascii
```
