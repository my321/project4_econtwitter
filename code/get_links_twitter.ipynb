##### LOOP THROUGH LIST OF URLS TO GRAB ALL THE TWITTER USERNAME 

```python
twitter_usernames=[]
for link in links: 
    url = 'https://ideas.repec.org'+link 
    response = requests.get(url)
    page=response.text
    soup=BeautifulSoup(page, 'lxml')
    handle=soup.find_all('table')[0].find(class_="fa fa-twitter").next
    handle=handle.replace("@","")
    handle=handle.strip()
    twitter_usernames.append(handle)
    
```

```python
len(twitter_usernames)
```

##### SAVE LIST OF TWITTER USERNAMES

```python
with open('twitter_usernames.txt', 'w') as filehandle:
    json.dump(twitter_usernames, filehandle)
```
